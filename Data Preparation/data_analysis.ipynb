```python
import pandas as pd
```

```python
df = pd.read_parquet('hospital_data.parquet')
```

```python
df['Specialties'] = df['Specialties'].str.split(', ')
```

```python
df['Specialties'].explode().value_counts()
```

```python
pincode_df = pd.read_csv('india_pincodes.csv')
pincode_df.head()
```

```python
pincode_df = pincode_df[['postal code ', 'latitude', 'longitude']].astype({'postal code ': str}).rename(columns={'postal code ': 'Pincode'})
pincode_df
```

```python
main_df = df.merge(pincode_df, on='Pincode', how='left')
main_df
```

```python
main_df.dropna(subset=['latitude', 'longitude'], inplace=True)
```

```python
main_df.to_parquet('hospital_data_enriched.parquet', index=False)
```
